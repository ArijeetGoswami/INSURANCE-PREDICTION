```python
# @title Data preprocessing
```

```python
df1.head()
```

```python
#converting the text data to numerical value() for numberical value()
#encoding thhe categorical fetures
#for the sex column
df1.replace({'sex':{'male':0,'female':1}},inplace=True)
#for the smoker column
df1.replace({'smoker':{'yes':0,'no':1}},inplace=True)
#for the region column
df1.replace({'region':{'southeast':0,'southwest':1,'northeast':2,'northwest':3}},inplace=True)
```

```python
df1.head()
```

```python
#splitting the features and target
#first dropiing the column (charges)
x=df1.drop(columns='charges',axis=1)
#stroing the chagres column in y
y=df1['charges']
```

```python
print(x)
```

```python
print(y)
```

```python
# @title for the ml model
```

```python
#spltting the data into train ,test and split   (arrays)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2) #split the data into two manner(random)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#model training
#load the LINEAR REGIRESSION MODEL
LR=LinearRegression()
```

```python
#noe fit the model in the training data
LR.fit(x_train,y_train)
```

```python
# @title MODEL EVALIUTAION
```

```python
#PREDICTION ON TRAIN DATA
TRAINING_DATA_PREDICT=LR.predict(x_train)
```

```python
# now r.square value() for training data
r2_train=metrics.r2_score(y_train,TRAINING_DATA_PREDICT)
print("r value : ",r2_train)
```

```python
# @title upto 75% for the training data
```

```python

#evalutaion on testing model
TRAINING_DATA_PREDICT=LR.predict(x_test)
R2_TEST=metrics.r2_score(y_test,TRAINING_DATA_PREDICT)
print("r value : ",R2_TEST)
```

```python
# @title APPROX SAME FOR BOTH TARIN AND TEST DATA  (SHOULD BE ALMOST EQUAL)
```

```python
#FINAL STEP TO BUILD THE PREDCTION SYSTEM

# INPUT_DATA=(37,male,29.83,2,no,northeast,6406.4107)
#change as per the values
input_data= (37,0,29.83,2,1,2)

#chanigin the data-type to array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the array
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

prediction=LR.predict(input_data_reshaped)

```
